Task 4 - 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample

# Parameters
f1 = 5  # Frequency of sine wave
f2 = 15  # Frequency of cosine wave
t = np.linspace(0, 1, 1000, endpoint=False)  # Time vector
mixed_signal = np.sin(2 * np.pi * f1 * t) + np.cos(2 * np.pi * f2 * t) # Mixed signal

fs = 20 # Sampling frequency

# Sampling the signal
t_samples = np.arange(0, 1, 1 / fs)  # Sampling time points
sampled_signal = np.sin(2 * np.pi * f1 * t_samples) + np.cos(2 * np.pi * f2 * t_samples)

# Reconstructing the signal (optional)
reconstructed_signal = resample(sampled_signal, len(t))

# Plotting
plt.figure(figsize=(12, 6))

# Original and sampled signals
plt.subplot(2, 1, 1)
plt.plot(t, mixed_signal, label='Original Signal')
plt.stem(t_samples, sampled_signal, linefmt='r-', markerfmt='ro', basefmt=" ", label='Sampled Signal')
plt.title('Mixed Signal (Sine + Cosine) and Sampled Points')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, mixed_signal, label='Original Signal')
plt.plot(t, reconstructed_signal, '--', label='Reconstructed Signal')
plt.title('Signal Reconstruction')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def quantize(signal, levels):
    min_val, max_val = min(signal), max(signal)
    step_size = (max_val - min_val) / levels
    quantized_signal = np.round((signal - min_val) / step_size) * step_size + min_val
    return quantized_signal

# Example
original_signal = np.array([0.1, 0.5, 1.0, 1.5, 2.0])
quantized_signal = quantize(original_signal, levels=4)

print("Original Signal:", original_signal)
print("Quantized Signal:", quantized_signal)

# Element-wise quantization error
quantization_error_elements = original_signal - quantized_signal

# Mean Absolute Error (MAE)
mean_absolute_error = np.mean(np.abs(quantization_error_elements))

# Mean Squared Error (MSE)
mean_squared_error = np.mean(quantization_error_elements**2)

# Sum of Squared Error (SSE)
sum_squared_error = np.sum(quantization_error_elements**2)

print("Mean Absolute Quantization Error:", mean_absolute_error)
print("Mean Squared Quantization Error:", mean_squared_error)
print("Sum of Squared Quantization Error:", sum_squared_error)
```
