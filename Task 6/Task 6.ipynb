```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter
from statsmodels.tsa.arima.model import ARIMA

# Generate Random Walk Signal
N = 1000
steps = np.random.choice([-1, 1], size=N)
signal = np.cumsum(steps)

plt.figure(figsize=(12, 6))
plt.plot(signal)
plt.title("Random Walk")
plt.xlabel("Step")
plt.ylabel("Position")
plt.grid(True)
plt.show()

# AR(1)
ar1_model = ARIMA(signal, order=(1, 0, 0)) # ARIMA(1,0,0) mathematically identical to AR(1)
ar1_model_fit = ar1_model.fit()
ar1_forecast = ar1_model_fit.predict(start=0, end=N-1)

plt.figure(figsize=(12, 6))
plt.plot(signal, label="Original")
plt.plot(ar1_forecast, label="AR(1) Fit", alpha=0.7)
plt.legend()
plt.title("AR(1) Model Fitting")
plt.xlabel("Sample")
plt.ylabel("Value")
plt.grid(True)
plt.show()

# ARIMA(1,1,0)
trend = signal

arima110_model = ARIMA(trend, order=(1, 1, 0))
arima110_model_fit = arima110_model.fit()

arima110_forecast = arima110_model_fit.predict(start=0, end=N-1)

plt.figure(figsize=(12, 6))
plt.plot(trend, label="Original")
plt.plot(arima110_forecast, label="ARIMA(1,1,0) Fit", alpha=0.7)
plt.legend()
plt.title("ARIMA Model Fitting")
plt.xlabel("Sample")
plt.ylabel("Value")
plt.grid(True)
plt.show()

# ARIMA(2,1,2)
trend = signal

arima212_model = ARIMA(trend, order=(2, 1, 2))
arima212_model_fit = arima212_model.fit()

arima212_forecast = arima212_model_fit.predict(start=0, end=N-1)

plt.figure(figsize=(12, 6))
plt.plot(trend, label="Original")
plt.plot(arima212_forecast, label="ARIMA(2,1,2) Fit", alpha=0.7)
plt.legend()
plt.title("ARIMA Model Fitting")
plt.xlabel("Sample")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
