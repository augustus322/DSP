```python
import numpy as np
import matplotlib.pyplot as plt

# Given frequency domain data
X = np.array([6, 4, 4, 5, 3, 4, 5, 0, 0, 0, 0])

# Values of N to test
N_values = [4, 8, 11, 16]

plt.figure(figsize=(12, 8))

for i, N in enumerate(N_values, 1):
    # Create time indices (K matrix)
    K = np.arange(N).reshape(-1, 1)
    
    # Create frequency indices
    n = np.arange(N)
    
    # Create the IDFT matrix W
    W = np.exp(2j * np.pi * K * n / N) / N
    
    # Zero-pad or truncate X to length N
    X_padded = np.zeros(N, dtype=complex)
    X_padded[:len(X)] = X[:N]
    
    # Compute IDFT
    x = np.real(W @ X_padded)
    
    # Plot the synthesized signal
    plt.subplot(2, 2, i)
    plt.stem(x)
    plt.title(f'Synthesized Signal (N={N})')
    plt.xlabel('Time index')
    plt.ylabel('Amplitude')
    plt.grid(True)
    
    # Display matrices
    print(f"\nFor N={N}:")
    print("\nW matrix (IDFT matrix):")
    print(np.round(W, 3))
    print("\nK matrix (time indices):")
    print(K)

plt.tight_layout()
plt.show()
```
